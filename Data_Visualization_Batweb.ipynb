```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "data_vis.csv"
df = pd.read_csv(file_path)
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.head())
```

```python
# Data Visualization

# Bar Graph: Count of Wines by Country
plt.figure(figsize=(14, 8))
country_count = df['country'].value_counts().head(15)
country_count.plot(kind='bar', color='green')
plt.title('Top 15 Countries with Most Wines')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Count of Varieties
plt.figure(figsize=(14, 8))
variety_count = df['variety'].value_counts().head(15)
variety_count.plot(kind='bar', color='skyblue')
plt.title('Top 15 Wine Varieties')
plt.xlabel('Wine Variety')
plt.ylabel('Count')
plt.show()
```

```python
# Count of Wines by Province
plt.figure(figsize=(16, 8))
province_count = df['province'].value_counts().head(20)
province_count.plot(kind='bar', color='salmon')
plt.title('Top 20 Provinces with Most Wines')
plt.xlabel('Province')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Histogram: Distribution of Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True, color='orange')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pairplot for numerical columns
sns.pairplot(df[['points', 'price']])
plt.suptitle('Pairplot of Points and Price', y=1.02)
plt.show()
```

```python
# Pie Chart: Distribution of Varieties
plt.figure(figsize=(10, 10))
variety_distribution = df['variety'].value_counts().head(10)
plt.pie(variety_distribution, labels=variety_distribution.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Top 10 Wine Varieties Distribution')
plt.show()
```

```python
# Heatmap: Correlation Heatmap for selected columns
plt.figure(figsize=(10, 8))
selected_columns = ['points', 'price']
correlation_matrix_selected = df[selected_columns].corr()
sns.heatmap(correlation_matrix_selected, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap for Points and Price')
plt.show()
```

```python
#  Boxplot: Distribution of Points by Province
plt.figure(figsize=(14, 8))
sns.boxplot(x='province', y='points', data=df.head(500))  # Limiting to first 500 rows for better visibility
plt.title('Boxplot of Points by Province')
plt.xlabel('Province')
plt.ylabel('Points')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Scatter Plot: Points vs. Price
plt.figure(figsize=(10, 8))
sns.scatterplot(x='points', y='price', data=df.head(500), color='purple', alpha=0.7)  # Limiting to first 500 rows for better visibility
plt.title('Scatter Plot of Points vs. Price')
plt.xlabel('Points')
plt.ylabel('Price')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
#  Violin Plot: Distribution of Points by Country
plt.figure(figsize=(14, 8))
sns.violinplot(x='country', y='points', data=df.head(1000))  # Limiting to first 1000 rows for better visibility
plt.title('Violin Plot of Points by Country')
plt.xlabel('Country')
plt.ylabel('Points')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Pairplot: Pairwise Relationships for selected columns
sns.pairplot(df[['points', 'price', 'variety']], hue='variety', palette='Set1', height=4)
plt.suptitle('Pairplot of Points, Price, and Variety', y=1.02)
plt.show()
```

```python
# Barh Graph: Average Price by Variety
plt.figure(figsize=(12, 8))
average_price_by_variety = df.groupby('variety')['price'].mean().sort_values(ascending=False).head(15)
average_price_by_variety.plot(kind='barh', color='skyblue')
plt.title('Top 15 Varieties by Average Price')
plt.xlabel('Average Price')
plt.ylabel('Variety')
plt.show()
```

```python
#  Line Plot: Trends in Points Over Time (Assuming there's a 'year' column in the dataset)
if 'year' in df.columns:
    plt.figure(figsize=(12, 6))
    points_over_time = df.groupby('year')['points'].mean()
    points_over_time.plot(kind='line', marker='o', color='red')
    plt.title('Average Points Over Time')
    plt.xlabel('Year')
    plt.ylabel('Average Points')
    plt.show()
else:
    print("No 'year' column found in the dataset.")
```
